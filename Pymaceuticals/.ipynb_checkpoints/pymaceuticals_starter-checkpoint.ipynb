```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data_df = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_data_df = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
pymaceuticals_df = pd.merge(mouse_drug_data_df, clinical_trial_data_df, on='Mouse ID', how="left")

# Display the data table for preview
pymaceuticals_df.head()
```

Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
drug_grouped = pymaceuticals_df.groupby(["Drug", "Timepoint"])["Tumor Volume (mm3)"]
mean_tumor_vol = drug_grouped.mean()
# Convert to DataFrame
mean_tumor_vol_df= pd.DataFrame(mean_tumor_vol)
# Preview DataFrame
mean_tumor_vol_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
std_error_tumor_vol = drug_grouped.sem()
# Convert to DataFrame
std_error_tumor_vol_df= pd.DataFrame(std_error_tumor_vol)
# Preview DataFrame
std_error_tumor_vol_df.head()

```

```python
# Minor Data Munging to Re-Format the Data Frames
pymaceuticals_pivot_df = pymaceuticals_df.pivot_table(index="Timepoint",columns="Drug",values="Tumor Volume (mm3)")

# Preview that Reformatting worked
pymaceuticals_pivot_df.head()
```

```python
# Generate the Plot (with Error Bars)
drug_list = [("Capomulin", "red", "o"),("Infubinol", "blue", "^"),("Ketapril", "green", "s"),("Placebo","black","d")]
for drug,colors,marker in drug_list:
    std = std_error_tumor_vol[drug]
    tum_response_treatment_plt = plt.errorbar(pymaceuticals_pivot_df.index,pymaceuticals_pivot_df[drug],std,marker=marker,
                                       color=colors,ls='--',linewidth=1)
plt.legend(loc='best')
plt.title('Tumor Response to Treatment')
plt.xlabel('Time (Days)')
plt.ylabel('Tumor Volume (mm3)')
plt.grid()

# Save the Figure
plt.savefig("../Images/Tumor_response_to_treatment.png")
```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
met_resp = pymaceuticals_df.groupby(['Drug','Timepoint'])['Metastatic Sites']
met_resp_df = met_resp.mean()
# Convert to DataFrame
metastatic_response_df = pd.DataFrame(met_resp_df)
# Preview DataFrame
metastatic_response_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
met_std_err = met_resp.sem()

# Convert to DataFrame
met_std_err_df = pd.DataFrame(met_std_err)

# Preview DataFrame
met_std_err_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
metastatic_response_pivot = metastatic_response_df.pivot_table(index="Timepoint",columns="Drug",values="Metastatic Sites")
# Preview that Reformatting worked
metastatic_response_pivot.head()
```

```python
# Generate the Plot (with Error Bars)
for drug,colors,marker in drug_list:
    std = met_std_err[drug]
    met_response_treatment_plt = plt.errorbar(metastatic_response_pivot.index,metastatic_response_pivot[drug],std,marker=marker,
                                       color=colors,ls='--',linewidth=1)
plt.legend(loc='best')
plt.title('Metastatic Spread During Treatment')
plt.xlabel('Treatment Duration (Days)')
plt.ylabel('Met. Sites')
plt.grid()

# Save the Figure
plt.savefig("../Images/Metastic_Spread.png")

# Show figure
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mice_group = pymaceuticals_df.groupby(['Drug','Timepoint'])['Mouse ID']
mice_count = mice_group.count()

# Convert to DataFrame
mice_count_df = pd.DataFrame(mice_count)
mice_count_df=mice_count_df.rename(columns={'Mouse ID':'Mouse Count'})

# Preview DataFrame
mice_count_df.head()

```

```python
# Minor Data Munging to Re-Format the Data Frames
mice_count_pivot = mice_count_df.pivot_table(index="Timepoint",columns="Drug",values="Mouse Count")
# Preview the Data Frame
mice_count_pivot.head()
```

```python
# Generate the Plot (Accounting for percentages)
for drug,colors,marker in drug_list:
    survival_rate_plt = plt.plot(mice_count_pivot.index,(mice_count_pivot[drug]/mice_count_pivot[drug][0])*100,marker=marker,
                                       color=colors,ls='--',linewidth=1)
plt.legend(loc='best')
plt.title('Survival During Treatment')
plt.xlabel('Time (Days)')
plt.ylabel('Survival Rates(%)')
plt.grid()
# Save the Figure
plt.savefig("../Images/Survival_During_Treatment.png")
# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
percent_change = (pymaceuticals_pivot_df.iloc[-1]/(pymaceuticals_pivot_df.iloc[0])-1)*100
# Display the data to confirm
percent_change
```

```python
# Store all Relevant Percent Changes into a Tuple
drug = ["Capomulin", "Infubinol", "Ketapril", "Placebo"]
prct_change = [(percent_change[Drug]) for Drug in drug]


# Splice the data between passing and failing drugs
passing_failing = percent_change < 0

# Orient widths. Add labels, tick marks, etc. 
percent_change_plt = plt.bar(drug, prct_change,color=passing_failing.map({True:'green',False:'red'}),align='edge', width=-1)

plt.grid()
plt.ylim(-25,80)


# Use functions to label the percentages of changes
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        if height > 0:
            label_position = 2
        else:
            label_position = -8
        plt.text(rect.get_x() + rect.get_width()/2., label_position,
                '%d' % int(height)+'%',color='white',
                ha='center', va='bottom')


# Call functions to implement the function calls
autolabel(percent_change_plt)

# Make some labels.
plt.title("Tumor Change Over 45 Day Treatment")
plt.ylabel("% Tumor Volume Change")

# Save the Figure
plt.savefig("../Images/Tumor_Percent_Change.png")

# Show the Figure
plt.show()
```
